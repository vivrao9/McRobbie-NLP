```python
wordfreq={}

for speech in os.listdir(r'C:\Users\Vivek Rao\Desktop\IDS\Digital\McRobbie NLP\Speeches'):
    file_path = 'C:\\Users\\Vivek Rao\\Desktop\\IDS\\Digital\\McRobbie NLP\\Speeches\\' + speech
    f = open(file_path, "r", encoding="utf-8")
    f = f.readlines()
    for sentence in f:
        for sent in nltk.sent_tokenize(sentence):
            for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sent))):
                if hasattr(chunk, 'label'):
                    for c in chunk:
                        if c[0] not in wordfreq:
                            wordfreq[c[0]] = 0
                        wordfreq[c[0]] += 1 #print(chunk.label(), ' '.join(c[0] for c in chunk))
                        
{k: v for k, v in sorted(wordfreq.items(), reverse=True, key=lambda item: item[1])}
```
