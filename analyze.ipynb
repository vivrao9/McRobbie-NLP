```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.tokenize import regexp_tokenize
from nltk.corpus import stopwords
from nltk.corpus import wordnet
from nltk.corpus import webtext
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures
from nltk.collocations import TrigramCollocationFinder
from nltk.metrics import TrigramAssocMeasures
```

```python
speech = open('./SotU/2010.txt', 'r', encoding="utf-8")
```

```python
sotu = speech.read()
```

```python
#split text into sentences
sent_tokenize(sotu)
```

```python
#split text into words
sotu = regexp_tokenize(sotu, "[\w']+")

#set stopwords
english_stops = set(stopwords.words('english'))

#filter out stopwords
sotu = [word for word in sotu if word not in english_stops]
```

```python
#find words that are most likely to be paired together
bcf = BigramCollocationFinder.from_words(sotu)
#thesecond argument here outputs <n> number of pairs
bcf.nbest(BigramAssocMeasures.likelihood_ratio, 10)
```

```python
#find words that are likely to be paired together in threes (e.g. long term relationship)
tcf = TrigramCollocationFinder.from_words(sotu)
#remove trigrams that occured less than three times
tcf.apply_freq_filter(3)
tcf.nbest(TrigramAssocMeasures.likelihood_ratio, 10)
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
for word in sotu:
    word = stemmer.stem(word)
len(sotu)
```
